```python
from dotenv import load_dotenv
load_dotenv()
```

```python
# 1. 필요한 라이브러리를 임포트합니다
from langchain.agents import create_agent
from langchain_classic.chains.summarize import load_summarize_chain
```

```python
# 2. 뉴스 요약을 위해 OpenAI 에이전트를 생성합니다
agent = create_agent(
    model="openai:gpt-4.1-mini",
    tools=[],
    system_prompt=(
        "너는 뉴스 기사와 장문 문서를 요약하는 전문 에이전트다. "
        "문서의 핵심 사실, 주요 사건, 관련 인물·기관, 날짜·수치를 우선적으로 파악하라. "
        "중복 표현과 불필요한 수식은 제거하고, 핵심만 간결하게 정리하라. "
        "원문에 없는 내용을 추측하거나 보충하지 마라. "
        "출력은 반드시 한국어로 하라. "
        "다음 형식을 지켜라. "
        "1) 한 줄 요약 "
        "2) 핵심 내용 3개 bullet "
        "3) 중요 키워드 3~5개"
    ),
)
```

```python
# 3. 요약 체인 생성
summarize_chain = load_summarize_chain(agent, chain_type="map_reduce")

result = agent.invoke({
    "messages": [{"role": "user", "content": "안녕하세요" }]
})

print(result)
```
